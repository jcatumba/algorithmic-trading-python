```python
url = "https://api.mercadolibre.com/sites/MCO/search"
```

```python
offset = 0
results = []
while True:
    querystring = {"q":"rodillo+quita+pelos", "offset":offset}

    payload = ""
    response = requests.request("GET", url, data=payload, params=querystring)
    this_iter_results = json.loads(response.text)

    results += this_iter_results['results']
    
    if offset==0:
        print(this_iter_results['paging']['total'])

    offset += this_iter_results['paging']['limit']
    if (offset > this_iter_results['paging']['total']) or offset==1000:
        break
```
